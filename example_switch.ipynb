```python
import pickle
import networkx as nx
from utils.switch import simple_partition, godsil_mckay_switch
from utils.measures import are_cospectral, adjacency_spectrum
import matplotlib.pyplot as plt
```

```python
with open("./cospectrality_undirected/graph_1.pkl", "rb") as f:
    G = pickle.load(f)
```

```python
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, edge_color="gray")

plt.show()
```

```python
set1, set2 = simple_partition(G, n=2)

color_map = []
for node in G.nodes():
    if node in set1:  # C
        color_map.append("skyblue")  # first set color
    else:  # D
        color_map.append("lightgreen")  # second set color
```

```python
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color=color_map, edge_color="gray")
plt.show()
```

```python
h = godsil_mckay_switch(G, set1, set2)
pos = nx.planar_layout(h)
nx.draw(h, pos, with_labels=True, node_color=color_map, edge_color="gray")
plt.show()
```

```python
print(h.edges())
```

```python
print(are_cospectral(G, h))
```

```python
print(adjacency_spectrum(G))
print(adjacency_spectrum(h))
```

```python
GM = nx.algorithms.isomorphism.GraphMatcher

print(GM(G, h).is_isomorphic())
```
